```python
from bing_image_downloader import downloader
import os
import shutil
```

```python
output_folder = "dataset_nasi_kuning"
if os.path.exists(output_folder):
    shutil.rmtree(output_folder)  # Hapus folder jika sudah ada
```

```python
search_query = "nasi kuning"
downloader.download(search_query, limit=100, output_dir=output_folder, adult_filter_off=True, force_replace=False)
```

```python
import zipfile

# Membuat file ZIP
zip_file = "dataset_nasi_kuning.zip"
with zipfile.ZipFile(zip_file, 'w') as zipf:
    for root, dirs, files in os.walk(output_folder):
        for file in files:
            zipf.write(os.path.join(root, file), os.path.relpath(os.path.join(root, file), output_folder))

print(f"File ZIP dibuat: {zip_file}")
```
